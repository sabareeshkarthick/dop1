### problem 3

Lets convert maximum problem to minimum problem by changing the sign,<br/>
$$min f=-x_1 x_2-x_2 x_3-x_1 x_3$$<br/>
$$h=x_1+x_2+x_3-3=0$$<br/>
REDUCED GRADIENT:<br/>
Consider $x_1$ as state variable and $x_2$ and $x_3$ as decision variable,<br/>
$\frac{\partial f}{\partial d}=\begin{bmatrix}-x_1-x_3 \\ -x_1-x_2 \end{bmatrix}$<br/>
$\frac{\partial f}{\partial s}=-x_2-x_3$<br/>
$\frac{\partial h}{\partial d}=\begin{bmatrix}1 \\ 1 \end{bmatrix}$<br/>
$\frac{\partial h}{\partial s}=1$<br/>
$\frac{ds}{dd}=-(\frac{\partial h}{\partial s})^{-1}(\frac{\partial h}{\partial d})=\begin{bmatrix}-1\\ -1 \end{bmatrix}$<br/>
$\frac{df}{dd}=\frac{\partial f}{\partial d}-\frac{\partial f}{\partial s}(\frac{\partial h}{\partial s})^{-1}(\frac{\partial h}{\partial d})=\begin{bmatrix}-x_1-x_3 \\ -x_1-x_2\end{bmatrix}-(-x_2-x_3)\begin{bmatrix} 1 \\ 1 \end{bmatrix}=\begin{bmatrix}-x_1+x_2 \\ -x_1+x_3\end{bmatrix}=\begin{bmatrix} 0\\ 0 \end{bmatrix}$<br/>
Substitute $x_1=x_2=x_3$ in constrain h ,we get,<br/>
$3 x_1-3=0$<br/>
$\implies x_1=3/3=1$<br/>
$\implies x_1=x_2=x_3=1$<br/>
Second order condition,<br/>
$\frac{\partial ^2 h}{\partial d^2}=\begin{bmatrix}0&0 \\ 0&0 \end{bmatrix}$<br/>
$\frac{\partial ^2 h}{\partial s^2}=0$<br/>
$\frac{\partial ^2 h}{\partial s \partial d}=\begin{bmatrix}0&0 \end{bmatrix}$<br/>
$\frac{\partial ^2 h}{\partial d \partial s}=\begin{bmatrix}0 \\ 0 \end{bmatrix}$<br/>
$\frac{d^2 h}{dd^2}=\begin{bmatrix}I & (\frac{ds}{dd})^T \end{bmatrix}\begin{bmatrix}\frac{\partial ^2 h}{\partial d^2} & \frac{\partial ^2 h}{\partial d \partial s}\\ \frac{\partial ^2 h}{\partial s \partial d}&\frac{\partial ^2 h}{\partial s^2} \end{bmatrix} \begin{bmatrix} I \\ \frac{ds}{dd}\end{bmatrix}+\frac{\partial h}{\partial s}\frac{d^2s}{dd^2}=0$<br/>
since $\frac{d^2s}{dd^2}=0$,<br/>
$\frac{\partial ^2 f}{\partial d^2}=\begin{bmatrix} 0 & -1 \\-1 & 0 \end{bmatrix}$<br/>
$\frac{\partial ^2 f}{\partial s^2}=0$<br/>
$\frac{\partial ^2 f}{\partial s \partial d}=\begin{bmatrix} -1 &-1 \end{bmatrix}$<br/>
$\frac{\partial ^2 f}{\partial d \partial s}=\begin{bmatrix} -1 \\ -1 \end{bmatrix}$<br/>
$\frac{d^2 f}{dd^2}=\begin{bmatrix}I & (\frac{ds}{dd})^T \end{bmatrix}\begin{bmatrix}\frac{\partial ^2 h}{\partial d^2} & \frac{\partial ^2 h}{\partial d \partial s}\\ \frac{\partial ^2 h}{\partial s \partial d}&\frac{\partial ^2 h}{\partial s^2} \end{bmatrix} \begin{bmatrix} I \\ \frac{ds}{dd}\end{bmatrix}+\frac{\partial h}{\partial s}\frac{d^2s}{dd^2}$<br/>
$\frac{d^2 f}{dd^2}==\begin{bmatrix}1&-1&-1\end{bmatrix}\begin{bmatrix}0&-1&-1\\-1&0&-1\\-1&-1&0\end{bmatrix}\begin{bmatrix}1\\-1\\-1\end{bmatrix}=2>0$<br/>
$\implies$when $x_1=x_2=x_3=1$ we get the optimal maximum solution 3.

LAGRANGE MULTIPLIERS:<br/>
$L=-x_1x_2-x_2x_3-x_1x_3+\mu (x_1+x_2+x_3-3)$<br/>
$\frac{\partial L}{\partial x}=\begin{bmatrix}-x_2-x_3+\mu \\ -x_1-x_3+\mu \\ -x_2-x_3+\mu \end{bmatrix}=\begin{bmatrix}0\\ 0 \\0 \end{bmatrix}$<br/>
$\frac{\partial L}{\partial \mu}=x_1+x_2+x_3-3=0$<br/>
Substitute$x_1=x_2=x_3=-\frac{\mu}{2}$ in the constrain ,then $x_1=x_2=x_3=1$and $\mu=-2$,<br/>
Second order condition,<br/>
$L_{xx}=\begin{bmatrix}0&-1&-1\\-1&0&-1\\-1&-1&0\end{bmatrix}$<br/>
$\frac{\partial h}{\partial x}=\begin{bmatrix}1\\1\\1\end{bmatrix}$<br/>
$(\frac{\partial h}{\partial x})^T dx=\begin{bmatrix}1&1&1\end{bmatrix}\begin{bmatrix}dx_1\\dx_2\\dx_3\end{bmatrix}=dx_1+dx_2+dx_3=0$<br/>
$dx^TL_{xx}dx=\begin{bmatrix}dx_1&dx_2&dx_3\end{bmatrix}\begin{bmatrix}0&-1&-1\\-1&0&-1\\-1&-1&0\end{bmatrix}\begin{bmatrix}dx_1\\dx_2\\dx_3\end{bmatrix}=-2dx_1dx_2-2dx_2dx_3-2dx_1dx_3$<br/>
Substitute $dx_1+dx_2+dx_3=0$,<br/>
$dx^TL_{xx}dx=2(dx^2_1+dx_1dx_2+dx_2^2)=2((dx_1+(1/2)dx_2)^2+(3/4)dx_2^2)>0$<br/>
$\implies x_1=x_2=x_3=1$ and $\mu=-2$ has maximum optimal solution which is 3.

```python
import numpy as np
def f(x1,x2,x3):
    return x1**2+x2**2+x3**2
def f_x(x1,x2,x3):
    f_s=np.vstack((2*x2,2*x3))
    f_d=np.array([2*x1],dtype=float)
    return f_s,f_d
def st(x1,x2,x3):
    h1 = x1**2/4+x2**2/5+x3**2/25-1
    h2 = x1+x2-x3
    return h1,h2
def h_x(x1, x2, x3):
    h_s=np.vstack((np.hstack(((2/5)*x2,(2/25)*x3)),np.array([[1,-1]],dtype=float)))
    h_si=np.linalg.inv(h_s)
    h_d=np.vstack(((1/2)*x1,1.))
    return h_si,h_d,h_s
def fd(x1,x2,x3):
    f_s,f_d=f_x(x1, x2, x3)
    h_si,h_d,_=h_x(x1, x2, x3)
    return f_d - np.matmul(f_s.T, np.matmul(h_si, h_d))
def out(x1,x2,x3):
    er=1e-3
    h1,h2=st(x1,x2,x3)
    h=np.vstack((h1,h2))
    h_n=np.linalg.norm(h)
    while h_n>=er:
        dh_i,_,dh=h_x(x1,x2,x3)
        k=1
        ds=np.matmul(dh_i, h)
        x2=x2-ds[0]
        x3=x3-ds[1]
        h1,h2=st(x1, x2, x3)
        h=np.vstack((h1, h2))
        h_n=np.linalg.norm(h)
    return x1,x2,x3, h_n
def l_s(x1,x2,x3):
    a = 1.  
    df = fd(x1, x2, x3)
    t = lambda a,x1,x2,x3,df:f(x1,x2,x3)-a*0.3*np.matmul(df,df.T) 
    def f_a(x1,x2,x3,a):
        df=fd(x1,x2,x3)
        h_si,h_d,_=h_x(x1, x2, x3)
        x1=x1-a*df.flatten()
        ds=np.matmul(np.matmul(h_si, h_d), df.T).flatten()
        x2=x2+a*ds[0]
        x3=x3+a*ds[1]
        return f(x1,x2,x3)
    while t(a,x1,x2,x3,df)<f_a(x1,x2,x3,a):
        a=0.5*a
        df=fd(x1,x2,x3)
    return a
r=1e-3 
x1=np.array([1.0],dtype=float)
x2=np.array([2.0],dtype=float)
x3=np.array([3.0],dtype=float)
i=0 
x1,x2,x3,_=out(x1,x2,x3)
f_d_n=np.linalg.norm(fd(x1, x2, x3))
while f_d_n>=r: 
    a=l_s(x1, x2, x3)
    x1=x1-a*fd(x1,x2,x3).flatten()
    h_si,h_d,dh=h_x(x1, x2, x3)
    ds=a*np.matmul(np.matmul(h_si, h_d), fd(x1,x2,x3).T).flatten()
    x2=x2+ds[0]
    x3=x3+ds[1]
    x1,x2,x3,_=out(x1,x2,x3)
    f_d_n=np.linalg.norm(fd(x1, x2, x3))
print('x1,x2,x3=({},{},{})'.format(x1,x2,x3))
```
